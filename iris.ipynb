```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from ggplot import * 
```

```python
import seaborn as sb
```

```python
x = load_iris()
```

```python
iris = pd.DataFrame(data= np.c_[x['data'], x['target']],
                 columns= list(x['feature_names']) + ['target'])
iris.head(5)
```

```python
iris.shape
```

```python
iris.describe()
```

```python
ggplot(aes('sepal length (cm)', 'sepal width (cm)', color='target'), data=iris) +\
    geom_point(size=30)
```

```python
sb.palplot(sb.cubehelix_palette(3, start=2.81, rot=-.2))
sb.set_palette(sb.cubehelix_palette(3, start=2.81, rot=-.2))
```

```python
sb.pairplot(data=iris, 
            hue='target', 
            vars=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
           )
```

```python
sb.boxplot(x='target', y='sepal length (cm)', data=iris)
```
